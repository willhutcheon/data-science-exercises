```python
import pandas as pd
import seaborn as sns
iris = sns.load_dataset('iris')
iris_df = pd.DataFrame(iris)
#iris_df.head()
sepal_data = iris_df[['sepal_length', 'sepal_width']]
covariance_matrix = sepal_data.cov()
print("Covariance Matrix")
print(covariance_matrix)
print()
correlation_matrix = sepal_data.corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

The covariance relationship between 'sepal_length' and 'sepal_width', with a value of -0.11757, displays a weak negative covariance, indicating that there is a slight tendency for sepal length to decrease as sepal width increases, and vice versa. However, the magnitude of this covariance is small, indicating that the relationship is weak.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
iris_df = pd.DataFrame(iris)
sepal_data = iris_df[['sepal_length', 'sepal_width']]
correlation_coefficient = sepal_data.corr().iloc[0, 1]
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris_df)
plt.text(4.5, 4.0, f'Correlation Coefficient: {correlation_coefficient:.2f}', fontsize=12)
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.show()
```
